## Ano 2020

```python
df3_2020 = df3.drop(columns=["2021","2019"])
df3_2020['Date'] =  df3_2020['Date'].apply(lambda x: x.replace(year = 2020))
df3_2020.columns=['Date', 'Travelers']
df3_2020 = df3_2020.set_index('Date')
df3_2020.head()
```

## Ano 2021

```python
df3_2021 = df3.drop(columns=["2020","2019"])
df3_2021['Date'] =  df3_2021['Date'].apply(lambda x: x.replace(year = 2021))
df3_2021.columns=['Date', 'Travelers']
df3_2021 = df3_2021.set_index('Date')
df3_2021.head()
```

```python
df3 = pd.concat([df3_2020,df3_2021])
#pd.set_option('display.max_rows', None)
df3 = df3.dropna()
df3
```

criar dataframes com colunas identificadoras de que pais pertencem

```python
df_d = []
i=0
for d in dataframes:
    d.columns=colunas[i]
    d = d.drop(columns = d.columns[[0,1,2]], axis =1)
    df_d.append(d)
    i=i+1
#df_d[5]
```

colocar as datas no formato corret

```python
for d in df_d:
    d["date"] = pd.to_datetime(d["date"], format = "%Y-%m-%d")
```

tornar datas como indices

```python
df_index =[]
```

```python
for d in df_d:
    df_index.append(d.set_index('date'))
```

Concatenar toos os dataframes e eliminar todas as colunas que contenham mais do que 60 valores nulos. Para esta eliminação não se considera as colunas relativas às vacinas, dado que é espectável que contenham grande número de valores nulos, pois só começaram a ser administradas muito tempo depois da data de ínicio do dataset

```python
result = pd.concat(df_index, axis=1)
nan_values = result.isnull().sum()
for j in nan_values.index:
    if nan_values[j] >60 and not j.find('people_vaccinated') and not j.find('people_fully_vaccinated_') and not j.find('new_vaccinations_') and not j.find('people_vaccinated_per_hundred_') and not j.find('people_fully_vaccinated_per_hundred_'):
        result = result.drop(columns = j, axis =1)
result.head()
```
