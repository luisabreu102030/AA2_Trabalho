Converção da variável "Date" para datetime de modo a ser possível usar pd.Grouper no groupby.

```python
df_covid["Date"] = pd.to_datetime(df_covid["Date"],format='%Y/%m/%d')
df_covid.dtypes
```

Agregação do dataset na passagem de registo diário para registo semanal usando o dicionário anteriormente criado para definir os diferentes métodos de agregação para as diversas colunas. Foi também escolhida como frequência de agregação o "W-SUN" de modo a que as datas originadas para a agregação coincidissem com as datas provenientes do dataset dos Estados Unidos. 

```python
df_covid = df_covid.groupby([pd.Grouper(key='Date', freq='W-SUN')]).agg(agg_dict).reset_index().sort_values('Date')
df_covid
```

Leitura do ficheiro CSV que contém os dados relativos a mortes semanais por doenças pulmonares nos Estados Unidos

```python
dataset = pd.read_csv('Datasets_utilizados/datasets_auxiliares/usa_dataset.csv', index_col=0)
dataset
```

Remoção de features irrelevantes, ou com informação facilmente obtida a partir de outras features do dataset. Features temporais já se encontravam no dataset relativo ao covid.

```python
dataset = dataset.drop(columns=["Max_Temp",	"Min_Temp","Temperature","Precipitation","Wind_Speed","Wind_Direction","Visibility",
                                "Cloud_Cover","Relative_Humidity",	"Rain",	"Clear"	,"Partially_cloudy","Nr People staying home",
                                "Total Deaths","Percent of Expected Deaths","Pneumonia and COVID-19 Deaths","Pneumonia, Influenza, or COVID-19 Deaths"])
		
```

Converção da variável "Date" para datetime de modo a ser possível ser efetuado o merge do dataset relativo a doenças pulmonares nos Estados Unidos com o dataset relativo ao covid em Portugal.

```python
dataset = dataset.reset_index().rename(columns={dataset.index.name:'Date'})
dataset["Date"] = pd.to_datetime(dataset["Date"],format='%Y/%m/%d')
dataset
```

```python
dataset.dtypes
```

```python
df_covid.dtypes
```

Efetuada a junção dos dois datasets em um só, usando como keys as datas de ambos.

```python
joined = pd.merge(df_covid, dataset, left_on="Date", right_on="Date", how = "left") #left_on="Date", right_on="Date"
joined
```

Verificação da existência de algum NA depois de efetuada a junção dos datasets. Como é possível verificar, existem alguns.

```python
joined.isna().sum()
```

Depois de verificado que esses NA encontram-se apenas na ultima observação em algumas das features, optou-se por fazer a sua interpolação, dando assim origem ao dataset semanal final.

```python
joined = joined.interpolate()
joined
```

```python
joined.to_csv("covid_&_influenza.csv")
```
