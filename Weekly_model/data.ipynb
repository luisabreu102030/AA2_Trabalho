```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Datasets/Provisional_COVID-19_Death_Counts_by_Week_Ending_Date_and_State.csv", sep = ",")
```

```python
data
```

```python
data1 = data.loc[(data["State"] == "United States") & (data["Group"] == "By Week")]
```

```python
data1
```

```python
data2 = data1.drop(columns=["Data as of", "Group", "Year", "Month", "Week Ending Date", "State", "Footnote", "Pneumonia and COVID-19 Deaths", "Pneumonia, Influenza, or COVID-19 Deaths","Percent of Expected Deaths"])
```

```python
data2.dtypes
```

```python
data2 = data2.astype({'MMWR Week': 'int64', 'COVID-19 Deaths': 'int64', 'Total Deaths': 'int64', 'Pneumonia Deaths': 'int64','Influenza Deaths': 'int64'})
```

```python
data2['Start Date'] = pd.to_datetime(data2["Start Date"],format='%m/%d/%Y') #, format='%Y%m%d', errors='coerce'
```

```python
data2['End Date'] = pd.to_datetime(data2["End Date"],format='%m/%d/%Y')
```

```python
port1 = port.loc[:,["data", "obitos"]]
```

```python
port1['data'] = pd.to_datetime(port1['data'],format="%d-%m-%Y")
```

```python
df = port1.groupby([pd.Grouper(key='data', freq='W-SAT')])['obitos'].sum().reset_index().sort_values('data')
```

```python
df
```

```python
df = df.set_index('data')
data2 = data2.set_index('End Date')
```

```python
final_data = pd.merge(df, data2, left_index=True, right_index=True, how = "outer")
```

```python
final_data = final_data.fillna(0)
```

```python
final_data = final_data.astype({'MMWR Week': 'int64', 'COVID-19 Deaths': 'int64', 'Total Deaths': 'int64', 'Pneumonia Deaths': 'int64','Influenza Deaths': 'int64', "obitos": "int64"})
```
