###### Leitura dos datasets com dados climatéricos para os Estados Unidos e junção dos mesmos num só

```python
temp1 = pd.read_csv("Datasets/temp_usa_01012020_01092020.csv", sep = ",")
temp2 = pd.read_csv("Datasets/temp_usa_02092020_31042021.csv", sep = ",")

df_temp = temp1.append(temp2, ignore_index = True)
```

```python
df_temp
```

###### Funções usadas para efetuar o one-hot-encoding ao atributo Conditions através de aplicação da função apply().

```python
def condition_ohe_rain(condition):
    res =0
    if(condition == 'Rain' or condition == 'Rain, Partially cloudy'):
        res=1

    return res

def condition_ohe_clear(condition):
    res =0
    if(condition == 'Clear'):
        res=1

    return res


def condition_ohe_cloudy(condition):
    res =0
    if(condition == 'Rain, Partially cloudy' or condition == 'Partially cloudy' or condition == 'Overcast'):
        res=1

    return res
```

###### Aplicação das funções de one-hot-encoding e eliminação de atributos com `missing values`

```python
###### Aplicação das funções de one-hot-encoding e eliminação de atributos com `missing values`
df_temp1 = df_temp.copy()
df_temp1 = df_temp1.drop(columns=['Name','Wind Chill','Heat Index', 'Wind Gust', 'Snow', 'Snow Depth'])
#print(df_aux.shape)

#rename columns
df_temp1.columns = ['Date', 'Max_Temp','Min_Temp','Temperature','Precipitation','Wind_Speed','Wind_Direction','Visibility','Cloud_Cover','Relative_Humidity','Conditions']

#one hot encoding weather conditions
df_temp1['Rain'] = np.nan
df_temp1['Clear'] = np.nan
df_temp1['Partially_cloudy'] = np.nan
    
df_temp1['Rain'] = df_temp1.apply(lambda x: condition_ohe_rain(x['Conditions']), axis=1)
df_temp1['Clear'] = df_temp1.apply(lambda x: condition_ohe_clear(x['Conditions']), axis=1)
df_temp1['Partially_cloudy'] = df_temp1.apply(lambda x: condition_ohe_cloudy(x['Conditions']), axis=1)

df_temp1 = df_temp1.drop(columns='Conditions')

df_temp1["Date"] = pd.to_datetime(df_temp1['Date'],format="%m/%d/%Y")
```

```python
df_temp1
```

###### Passagem do dataset de dias para semanas através do `group by`, obtendo a média de cada atributo

```python
df_temp2 = df_temp1.groupby([pd.Grouper(key='Date', freq='W')]).mean()
```
